데이터 프레임의 추가와 합치기

```python
# 행(수직)으로 합치기 :append(),concat()
score_table1 = {'영어': [10,20,40,50,50],
                '수학': [70,80,90,30,20]}
score_table2 = {'영어': [20,30,50,60,70],
                '수학': [80,90,80,40,30]}
df1 = pd.DataFrame(score_table1)
df2 = pd.DataFrame(score_table2)

# 추가 : new_dataframe = old_dataframe1.append(old_dataframe2)
new_df1 = df1.append(df2)  # 공통된 열을 사용하여 행을 추가
new_df2 = df1.append(df2,ignore_index=True)
new_df3 = df1.append(df1,ignore_index=True)
print(new_df2)
```

```python
pd.concat([df1,df2],axis = 0,ignore_index=True)
```

```python
# 열로 합치기 , merge(),join(),concat()

# concat()
# pd.concat(objs, axis=0, join='outer', join_axes=None,
# ignore_index=False,keys=None, levels=None, names=None, 
# verify_integrity=False, copy=True)
pd.concat([df1,df2],axis = 1)
```

```python
# merge()
# pd.merge(left, right, how='inner', on=None, left_on=None,
#          right_on=None,left_index=False, right_index=False,
#          sort=True,suffixes=('_x', '_y'), copy=True, 
#          indicator=False,validate=None)

dests = pd.read_csv('dest.csv')
print(dests)
tips = pd.read_csv('tips.csv')
print(tips)

# 내부조인 : 공통조건에 만족되는 부분만 조인된다
# 공통조건 = EmpNr
pd.merge(dests,tips, how='inner', on='EmpNr')

# 외부조인 : 공통조건 관계 없이 모두 조인된다
# 
pd.merge(dests,tips, how='outer')
```

```python
# join(other, on, how, lsuffix, rsuffix, sort)
dests.join(tips,lsuffix='_Dest',rsuffix='_Tips')
```

```python
# merge()활용
employee = [{'empno':1, 'ename':'kim', 'dept':1},
            {'empno':2, 'ename':'lee', 'dept':2}, 
            {'empno':3, 'ename':'park', 'dept':1}, 
            {'empno':4, 'ename':'song', 'dept':3},
            {'empno':5, 'ename':'min', 'dept':2} ]

dept=[{'dept':1, 'deptname':'관리직'},
      {'dept':2, 'deptname':'영업직'},
      {'dept':3, 'deptname':'개발직'} ]

info =[{'empno':1, 'addr':'서울시','phone':'010-1111-1111'},
       {'empno':3, 'addr':'부산시','phone':'010-2222-2222'}, 
       {'empno':2, 'addr':'광주시','phone':'010-3333-3333'}, 
       {'empno':5, 'addr':'광주시','phone':'010-4444-4444'},
       {'empno':4, 'addr':'광주시','phone':'010-5555-5555'} ]

emp = pd.DataFrame(employee)
dept = pd.DataFrame(dept)
info = pd.DataFrame(info)
print(emp)
print(dept)
print(info)
```

```python
# 내부 조인
m = pd.merge(emp,dept,how='inner',on='dept')

# drop(labels, axis, index, columns, level, inplace, errors)
m.drop(columns = ['dept'], inplace = True) # 'dept'컬럼삭제
# m.drop(index = [1,3],inplace = True) # 1,3 행을 삭제
# m
```

```python
m2 = pd.merge(m,info,how='inner',on='empno')
m2.drop(columns=['empno'], inplace = True)
m2
```

```python
# contains() 메소드를 사용하여 조건 검색, 행을 추출
sr = m2['ename'].str.contains('park')
type(sr)
print(sr)
m2[1:2] # 행으로 접근
print(m2[sr])

# 'deptname' = 영업직 인 행만 추출
print(m2[m2['deptname'].str.contains('영업직')])

# 'addr' = 광주시 인 행만 추출
print(m2[m2['addr'].str.contains('광주시')])
```

```python
# Series 객체의 통계 메소드 
score_table1 = {'영어': [10,20,40,50,50],
                '수학': [70,80,90,30,20]}
df1 = pd.DataFrame(score_table1)
sr = df1['영어']
type(sr)
# sum,mean,median,min,max
print('sum:',sr.sum())
print('mean:',sr.mean())
print('median:',sr.median())
print('min:',sr.min())
print('max:',sr.max())
print('describe:',sr.describe())
```
